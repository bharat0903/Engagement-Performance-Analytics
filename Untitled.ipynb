```python
import pandas as pd

users = pd.read_csv("edtech_users.csv")
eng = pd.read_csv("edtech_engagement.csv")
enr = pd.read_csv("edtech_enrollments.csv")
```

```python
df = users.merge(eng, on="user_id", how="left") \
          .merge(enr, on="user_id", how="left")
```

```python
df["viewed"] = df["course_viewed"].fillna(0)
df["demo"] = df["demo_attended"].fillna(0)
df["enrolled"] = df["enrollment_date"].notna().astype(int)
```

```python
funnel = df[["viewed","demo","enrolled"]].sum()
signups = len(df)
```

```python
import matplotlib.pyplot as plt

stages = ["Signups", "Viewed", "Demo", "Enrolled"]
counts = [signups, funnel["viewed"], funnel["demo"], funnel["enrolled"]]

plt.bar(stages, counts)
plt.title("User Funnel")
plt.show()
```
