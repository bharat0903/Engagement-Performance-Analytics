```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
creators = pd.read_csv("creators.csv")
users = pd.read_csv("users.csv")
videos = pd.read_csv("videos.csv")
video_impressions = pd.read_csv("video_impressions.csv")
video_engagements = pd.read_csv("video_engagements.csv")
```

```python
video_impressions["timestamp"] = pd.to_datetime(video_impressions["timestamp"])
video_engagements["timestamp"] = pd.to_datetime(video_engagements["timestamp"])
videos["upload_time"] = pd.to_datetime(videos["upload_time"])
```

```python
eng_full = (
    video_engagements
    .merge(videos, on="video_id", how="left")
    .merge(creators, on="creator_id", how="left")
    .merge(users, on="user_id", how="left")
)
```

```python
imp_full = (
    video_impressions
    .merge(videos, on="video_id", how="left")
    .merge(creators, on="creator_id", how="left")
    .merge(users, on="user_id", how="left")
)
```

```python
total_users = users["user_id"].nunique()
users_with_impressions = imp_full["user_id"].nunique()
users_with_engagements = eng_full["user_id"].nunique()

funnel_total = pd.DataFrame({
    "stage": ["Total Users", "Users Reached (Impressions)", "Users Engaged"],
    "count": [total_users, users_with_impressions, users_with_engagements]
})

funnel_total["conversion_rate"] = (
    funnel_total["count"] / funnel_total["count"].shift(1)
)
print(funnel_total)
```

```python
# Best & Worst Performing Regions
region_quality = (
    eng_full.groupby("region")
    .agg(avg_watch_time=("watch_time_sec", "mean"))
    .sort_values("avg_watch_time", ascending=False)
)

print(region_quality)
```

```python
# Biggest Drop-Off Points
funnel_total["dropoff"] = funnel_total["count"].shift(1) - funnel_total["count"]
funnel_total
```

```python
# Total Funnel Chart
plt.figure(figsize=(6,4))
plt.plot(funnel_total["stage"], funnel_total["count"], marker="o")
plt.title("ShareChat User Funnel")
plt.xlabel("Stage")
plt.ylabel("Users")
plt.grid()
plt.show()
```
